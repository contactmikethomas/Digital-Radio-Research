## Decompress and Cleanse Wikifiles

```python
import os
```

```python
pickle_list = []

input_dir = "/home/de-admin/Documents/open data/wikipedia category pickle/"
output_dir = "/home/de-admin/Documents/open data/wikipedia category final pickle/"

for pickle in os.listdir(input_dir):
    pickle_list.append(pickle)
```

```python
import pickle
import string
import mwparserfromhell
import lxml
from lxml import html, etree
```

```python
text_to_remove = ["#REDIRECT", "#redirect", "#Redirect"]

cat_list = []

i=0

for file in pickle_list:
    print("Unpickle data to process for category extraction. File %s" % str(i))
    pickle_part = pickle.load(open(input_dir+file, "rb" ))

    print("Category extraction preparation. File %s" % str(i))
    
    for doc in pickle_part:
        
        get_bit = 1
        cat_links = []
        
        try:
            doc_txt = html.fromstring(doc[1]).xpath('//text//text()')[0]

            for ttr in text_to_remove:
                if ttr in doc_txt:
                    get_bit = 0            
                elif "(disambiguation)" in doc[0]:
                    get_bit = 0
        except Exception as e:
            print(e)
            get_bit = 0

        if get_bit == 1:
            for link in mwparserfromhell.parse(html.fromstring(doc[1]).xpath('//text//text()')).filter_wikilinks():
                if "Category:" in link:
                    if link.find("|") == -1:
                        cat_links.append(link)
                    else:
                        cat_links.append(link[:link.find("|")]+link[link.find("]]"):])
            cat_list.append(['[['+doc[0]+']]',cat_links])
    
    i=i+1
```

```python
pickle_file = output_dir+"final category list"
pickle.dump(cat_list, open(pickle_file, 'wb'), pickle.HIGHEST_PROTOCOL)
```
